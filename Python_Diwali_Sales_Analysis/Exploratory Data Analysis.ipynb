# Exploratory Data Analysis


# Gender

```python
df.columns
```

```python
sns.countplot(x = 'Gender', data = df) # df is our dataframe
```

```python
ax = sns.countplot(x ='Gender', data= df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False).sum().sort_values(by='Amount', ascending=False)

ax = sns.barplot(x = 'Gender', y = 'Amount', data = sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
```

# From above grapgs we see that the most of the buyers are females and even then purchasing power of the females are greater than man too.

```python
# Total Amount V/S Gender wise Age Group
sales_age = df.groupby(['Gender','Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='Age Group', y='Amount', hue='Gender', data = sales_age)
for bars in ax.containers:
    ax.bar_label(bars)
```

#  State

```python
# Total number of orders from top 10 states
order_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
ax = sns.barplot(x='State', y='Orders', data= order_state)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# If I remove sort_values(by='Orders')
order_state = df.groupby(['State'], as_index=False)['Orders'].sum().head(10)
ax = sns.barplot(x='State', y='Orders', data= order_state)

sns.set(rc={'figure.figsize':(20, 5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total amount/sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x='State', y='Amount', data = sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

# Maritial Status

```python
sales_state = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
# hue='Gender' le gender ko husab le bar xuttauxa like Male and Female
ax = sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data =sales_state)
for bars in ax.containers:
    ax.bar_label(bars)
```

# Occupations

```python
sns.set(rc={'figure.figsize':(20,5)})
palette = sns.color_palette('husl', len(df['Occupation'].unique()))
ax = sns.countplot(x='Occupation', data = df,hue='Occupation', palette=palette)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
palette = sns.color_palette('husl', len(df['Occupation'].unique()))
# hue='Gender' le gender ko husab le bar xuttauxa like Male and Female
ax = sns.barplot(x='Occupation', y='Amount', hue='Occupation', data =sales_occupation, palette=palette)
# for bars in ax.containers:
#     ax.bar_label(bars)
```
